# 시계열 데이터 탐색 및 시각화
 - 시계열 데이터 시각화 소개 (선그래프, 박스 플롯, 자기상관 그래프)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### 선그래프

```python
#시계열 데이터 생성
np.random.seed(42)
dates = pd.date_range(start='2023-01-01', periods=1000, freq='D')
values = np.random.rand(1000)

# 데이터프레임 생성
data = pd.DataFrame({'Date' : dates,'Value' : values})

# 시계열 데이터 시각화
plt.plot(data['Date'], data['Value'])
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series Line Plot')
plt.show()
```

### 산점도 그래프

```python
#시계열 데이터 생성

date_temp = pd.date_range(start='2023-01-01', periods=1000, freq='D')
val_temp = np.random.randn(1000)

# 프레임 생성
data_pd = pd.DataFrame({
    'Date': date_temp,
    'Value': val_temp
})

print(data_pd)
```

#### 산점도 그래프 시각화

```python
plt.scatter(data_pd['Date'], data_pd['Value'])
plt.xticks()
```

#### 막대그래프

```python
dates_box_plot = pd.date_range(start='2023-01-01', periods=1000, freq='D')
val_box_plot = np.random.randn(1000)

box_plot_data = pd.DataFrame({
    'Date' : dates_box_plot,
    'Value' : val_box_plot
})

print(box_plot_data)
```

```python
plt.bar(box_plot_data['Date'], box_plot_data['Value'])
plt.xticks(rotation=-45)
plt.show()
```

#### 박스플롯 시각화

```python
box_date = pd.date_range(start='2023-01-01', periods=1000, freq='D')
box_val = np.random.rand(1000)

box_pd = pd.DataFrame({
    'Date' : box_date,
    'Value' : box_val
})

print(box_pd)
```

```python
plt.boxplot(box_pd['Value'])
plt.show()
```

### 히스토그램

```python
histogram_date = pd.date_range(start='2023-01-01', periods=1000, freq='D')
histogram_val = np.random.randn(1000)

histogram_data = pd.DataFrame({
    'Date' : histogram_date,
    'Value' : histogram_val
})

print(histogram_data)

plt.hist(histogram_data['Value'], bins=20)
plt.show()
```

### 히트맵 시각화

```python
import seaborn as sns
hit_date = pd.date_range(start='2023-01-01', periods=1000, freq='D')
hit_val = np.random.randn(1000)

hit_data = pd.DataFrame({
    'Date' : hit_date,
    'Value' : hit_val
})

print(hit_data)
```

```python
#데이터를 피벗 테이블 변환
pivot_table = hit_data.pivot(index= 'Date', columns='Value', values='Value')

#히트맵 시각화
plt.figure(figsize=(10,6))
sns.heatmap(pivot_table, cmap='YlOrRd', cbar=True)
plt.show()
```

피벗 테이블은 데이터를 특정 열을 기준으로 그룹화하여 요약하는 방법입니다.

예시
